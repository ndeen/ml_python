```python
def daily_ret_pct_diff(l):
    """
    calculates the daily return and pct diff between a value and its predecessor in a list.
    """

    diff_l = []  # create an empty list to hold the results
    pct_diff_l = []  # create an empty list to hold the results
    i = 0  # identify starting point

    # loop through list
    for val in l[
        i : len(l)
    ]:  # iterater from the starting point to the last value in list

        if i == 0:
            diff = 0
            pct_diff = 0

        elif i == 1:
            diff = l[i] - l[0]  # calculate diff b/w values
            pct_diff = diff / l[0]  # calcuate the pct diff

        else:
            diff = l[i] - l[i - 1]  # calculate diff b/w values
            pct_diff = diff / l[i]  # calculate the pct diff

        diff_l.append(diff)
        pct_diff_l.append(pct_diff)  # append results to container
        i += 1  # increment from starting point

    return diff_l, pct_diff_l  # return the results
```
