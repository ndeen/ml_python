```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(low=0, high=100, size=100)
```

```python
ax = plt.hist(x)
```

```python
y = np.random.randint(low=0, high=100, size=100)
```

```python
ax = plt.hist(y)
```

```python
plt.hist2d(x, y, bins=20)
plt.xlabel('low 0 high 100')
plt.ylabel('low 0 high 100')
plt.colorbar().set_label('Frequency of paired bin')
```
