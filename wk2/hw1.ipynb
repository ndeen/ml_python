```python
import os
```

```python
input_dir = os.path.abspath('')
ticker = 'DIS'
ticker_file = os.path.join(input_dir, ticker + '.csv')
```

```python
try:
    with open(ticker_file) as f:
        lines = f.read().splitlines()
    print('opened file for ticker: ', ticker)
    print(lines[0])
except Exception as e:
    print(e)
    print('failed to read stock data for ticker: ', ticker)
```

```python
# each line represents a row of data
# the first line represents the columns
cols = lines[0].split(',')
```

```python
date_l = []
year_l = []
weekday_l = []
ac_l = []

for row in lines[1:]:
    date = row.split(',')[0]
    date_l.append(date)
    
    year = row.split(',')[1]
    year_l.append(year)
    
    weekday = row.split(',')[4]
    weekday_l.append(weekday)
    
    ac = float(row.split(',')[12])
    ac_l.append(ac)
```

```python
ac_l
```

```python
def daily_ret_pct_diff(l):
    '''
    calculates the daily return and pct diff between a value and its predecessor in a list.
    '''
    
    diff_l = [] # create an empty list to hold the results
    pct_diff_l = [] # create an empty list to hold the results
    i = 0 # identify starting point
    
    # loop through list
    for val in l[i:len(ac_l)]: # iterater from the starting point to the last value in list
        print("current iteration:",i)
        
        if i == 0:
            diff = 0
            pct_diff = 0
        
        elif i == 1:
            diff = round(l[i] - l[0],2) # calculate diff b/w values
            pct_diff = round((diff / l[0]),2) # calcuate the pct diff
        
        else:
            diff = round((l[i] - l[i-1]),2) # calculate diff b/w values
            pct_diff = round((diff / l[i]),2) # calculate the pct diff
        
        diff_l.append(diff)
        pct_diff_l.append(pct_diff) # append results to container
        i+=1 # increment from starting point
        
        print("diff:",diff)
        print("pct diff:", pct_diff)
        print("next iteration:",i)
        print("\n")
    return diff_l, pct_diff_l # return the results
```

```python
diff_l, pct_diff_l = daily_ret_pct_diff(ac_l)
```
